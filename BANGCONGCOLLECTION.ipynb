# Viết hàm tự động chạy các sheet

```python
# Tạo một hàm đọc các sheet cho 1 file
total=[]
for i in sheet_name:
    df_data=pd.read_excel('BANG CONG T10-2022 AN HA.xlsx',sheet_name=i,skiprows=9,usecols="C,D,BQ,BR")
    df_data.columns=["Họ và tên","Bộ phận","HC","TC"]
    df_data["Ngày tháng"] = "Tháng10/2023"
    df_data["Xưởng"] = i
    df_data=df_data.dropna(axis=0)
    total.append(df_data)
    resultt=pd.concat(total)
```

# Viết hàm chạy tự động chạy các file

```python
# tạo đường dẫn đến thư mục chứa
path="C:/Users/Admin/Desktop/BANG CONG/BANG CONG/AN HA/"
```

```python
# sử dụng listdir duyệt qua các file trong thư mục đường dẫn path
os.listdir(path)
```

```python
sheet_name=["XUONG AH1","XUONG AH2","XUONG AH3"] 
```

```python
# Tạo hàm
frames=[]
for i in os.listdir(path):
    if i.endswith(".xlsx"):
        filepath = path + i
        for f in sheet_name:
            df1=pd.read_excel(filepath,sheet_name=f, skiprows=9,usecols="C,D,BQ,BR")
            df1.columns=["Họ và tên","Bộ phận","HC","TC"]
            df1["Tổng giờ"] = df1["HC"] + df1["TC"]
            df1["Xưởng"] = f[5:]
            df1["Ngày tháng"] = i[9:18]
            df1=df1.dropna(axis=0)
            frames.append(df1)
            result = pd.concat(frames)
```

```python
# Lưu file vào 1 file excel
result.to_excel("TH.xlsx")
```
